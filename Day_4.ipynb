```python
import numpy as np
from scipy.signal import convolve2d


def solve_my_puzzle(my_puzzle):
  lines = [list(line) for line in my_puzzle.strip().split("\n") if line]
  matrix = (np.array(lines)=='@').astype(int)
  print(matrix)

  kernal = np.array([[1,1,1],
                     [1,0,1],
                     [1,1,1]
                     ])
  total_removed_rolls = 0

  while True:
    count_adjacent_positions = convolve2d(matrix, kernal, mode='same')
    print(f"adjacent positions \n {count_adjacent_positions}")
    accessible_rolls = (matrix == 1) & (count_adjacent_positions < 4)
    current_accessible_rolls = np.sum(accessible_rolls)
    print(f"accessible_rolls \n {current_accessible_rolls}")

    if current_accessible_rolls == 0:
      break

    matrix[accessible_rolls] = 0
    total_removed_rolls += current_accessible_rolls
    print(f"total removable rolls is {total_removed_rolls}")



  return total_removed_rolls



extractor = solve_my_puzzle(my_puzzle)
print(extractor)
```

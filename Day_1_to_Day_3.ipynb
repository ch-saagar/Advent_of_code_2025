```python
def solve_my_puzzle(my_puzzle):
  total_joltage = 0

  for line in my_puzzle.split("\n"):
    line = line.strip()
    if not line:
      continue

    current_max = 0
    length_of_line = len(line)

    for first in range(length_of_line):
       for second in range(first+1, length_of_line):
          tens = int(line[first])
          ones = int(line[second])

          maximum_joltage = (tens * 10) + ones

          if maximum_joltage > current_max:
              current_max = maximum_joltage
          if current_max == 99:
            break
       if current_max == 99:
         break
    #print(f"current_maximum is {current_max}")
    total_joltage += current_max
  return total_joltage


extractor = solve_my_puzzle(my_puzzle)
print(extractor)
```

```python
def solve_my_puzzle(my_puzzle):
  total_joltage = 0
  total_length_req = 12

  for line in my_puzzle.split("\n"):
    line = line.strip()
    if not line:
      continue
    digits = [int(digit) for digit in line]
    buffer_length = len(digits) - total_length_req

    stack = []
    for digit in digits:
        while stack and buffer_length > 0 and stack[-1] < digit:
            stack.pop()
            buffer_length -= 1
        stack.append(digit)
    required_digits = stack[:total_length_req]
    total_joltage += int("".join(map(str, required_digits)))

  return total_joltage


extractor = solve_my_puzzle(my_puzzle)
print(extractor)
```
